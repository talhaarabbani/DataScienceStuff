```python
df['Dimensions - Height'] = df['Dimensions - Height'].astype(str).str.replace('ì', '')
df['Dimensions - Width'] = df['Dimensions - Width'].astype(str).str.replace('ì', '')

df['Dimensions - Height'] = df['Dimensions - Height'].astype(str).str.replace('"', '')
df['Dimensions - Width'] = df['Dimensions - Width'].astype(str).str.replace('"', '')

df['Dimensions - Height'] = pd.to_numeric(df['Dimensions - Height'], errors ='coerce')
df['Dimensions - Width'] = pd.to_numeric(df['Dimensions - Width'], errors= 'coerce')

```

```python
df['Dimensions - Width'] = df['Dimensions - Width'].astype(float)
df['Dimensions - Height'] = df['Dimensions - Height'].astype(float)
df['surfaceArea'] = df['Dimensions - Width'] * df['Dimensions - Height']
```
