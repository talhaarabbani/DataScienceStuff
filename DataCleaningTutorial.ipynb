```python
%%time

# using textblod for sentiment analysis first

pos_count = 0
pos_correct = 0
actual_NumOfPositive_lines = 0
neg_NumOfNegative_lines = 0


with open("/Users/TMRNEW/Documents/All_DS_Python_stuff/positive.txt","r") as f:
    for line in f.read().split('\n'):
        analysis = TextBlob(line)
        
        actual_NumOfPositive_lines +=1
        
        if analysis.sentiment.polarity >= 0.000001:
            pos_correct += 1
            pos_count +=1


            

neg_count = 0
neg_correct = 0

with open("/Users/TMRNEW/Documents/All_DS_Python_stuff/negative.txt","r") as f:
    for line in f.read().split('\n'):
        analysis = TextBlob(line)
        neg_NumOfNegative_lines +=1
        
        if analysis.sentiment.polarity <= 0.000001:
            neg_correct += 1
            neg_count +=1

print("Positive samples = {} samples".format(pos_count))
print("Negative samples = {} samples".format(neg_count))

print("actual positive samples = {}".format(actual_NumOfPositive_lines))
print("actual negative samples = {}".format(neg_NumOfNegative_lines))

```
